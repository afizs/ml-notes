## Hexbin Charts using Python

A hex bin chart is a 2D hexagonal binning plot of points which are an   
alternative of scatter plots when there are too many data points

```python
import numpy as np
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(8)

# Simulated Data
x = np.random.normal(0, 2, 1000000)
y = 3 * x + 5 * np.random.normal(0, 2, 1000000)

# Hexbin chart
fig, ax = plt.subplots()
ax.hexbin(x = x, y = y, gridsize = 15) # gridesize: number of hexagons
plt.show() 
```

## Border color and width for Hexagons

```python
import numpy as np
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(8)

# Simulated Data
x = np.random.normal(0, 2, 1000000)
y = 3 * x + 5 * np.random.normal(0, 2, 1000000)

# Hexbin chart
fig, ax = plt.subplots()
ax.hexbin(x = x, y = y, gridsize = 15,  # gridesize: number of hexagons
          edgecolor = "yellow", # border color
          linewidths = 1.5) # borderwidth
        
plt.show() 
```
